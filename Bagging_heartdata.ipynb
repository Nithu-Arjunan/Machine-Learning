Detecting outliers

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df[df.MaxHR>(df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df[df.FastingBS>(df.FastingBS.mean()+3*df.FastingBS.std())]
```

```python
## Label encoding the columns

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

col_to_encode = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

for col in col_to_encode:
    df3[col] = le.fit_transform(df3[col])
```

```python
df3
```

```python
X = df3.drop("HeartDisease",axis='columns')
y = df3['HeartDisease']
```

```python
# Scaling the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Using SVM to build the model- Standalone model

from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Building a bagging model using cross validation
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
scores = cross_val_score(bag_model, X, y, cv=5)
scores
```

As you can see above, using bagging in case of SVM doesn't make much difference in terms of model accuracy. Bagging is effective when we have high variance and instable model such as decision tree.


```python
# Using DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(random_state=0), X, y, cv=5)
scores.mean()
```

```python
# Building a model using bagging

bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
scores = cross_val_score(bag_model, X, y, cv=5)
scores
```

```python
scores.mean()
```

When used Bagging for decision tree classifier , accuracy has increased.
