```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
```

```python
df
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

Cross validation score for logistic regression

```python
l_scores = cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X, y,cv=10)
np.average(l_scores)
```

Cross validation score for SVM

```python
s_scores = cross_val_score(SVC(gamma='auto'), X, y,cv=10)
np.average(s_scores)
```

Cross Validation score for Decision Tree

```python
d_scores = cross_val_score(DecisionTreeClassifier(), X, y,cv=10)
np.average(d_scores)
```

Cross Validation score for Random Forest

```python
rf_scores = cross_val_score(RandomForestClassifier(n_estimators=40),X, y,cv=10)
np.average(rf_scores)
```

From the scores , the best model is SVM(Support Vector Machine)
