```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
```

```python
df
```

```python
df.experience.isna().sum()
```

```python
df.experience = df.experience.fillna('zero')
```

```python
import math
mean_test_score = math.floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mean_test_score)
```

```python
df
```

```python
# To convert word to number

!pip install word2number
from word2number import w2n # Import the w2n module

df['experience'] = df['experience'].astype(str).apply(lambda x: w2n.word_to_num(x) if x.lower() != 'nan' else np.nan)
```

```python
df
```

```python
model_linear = linear_model.LinearRegression()
model_linear.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
model_linear.coef_
```

```python
model_linear.intercept_
```

```python
model_linear.predict([[2, 9, 6]])
```

```python
model_linear.predict([[12, 10, 10]])
```

```python
# Saving the file into pickle
import pickle
with open('model_pickle', 'wb') as f:
    pickle.dump(model_linear, f)
```

```python
with open('model_pickle', 'rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[2, 9, 6]])
```
